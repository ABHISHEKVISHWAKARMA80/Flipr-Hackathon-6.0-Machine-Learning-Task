```python
std=np.std(X,axis=0)
u=np.mean(X,axis=0)
print(std)
print(u)
```

```python
X=(X-u)/std
print(X)
```

```python
ones=np.ones((X.shape[0],1))
X = np.hstack((ones,X))
print(X)
```

```python
twos=np.ones((x_test.shape[0],1))
x_test=np.hstack((twos,x_test))
print(x_test)
```

```python
def hypothesis(X,theta):
    return np.dot(X,theta)
def error(X,Y,theta):
    error=0.0
    m=X.shape[0]
    y_=hypothesis(X,theta)
    e=np.sum((Y-y_)**2)
    return e/m
def gradient(X,Y,theta):
    y_=hypothesis(X,theta)
    grad=np.dot(X.T,(y_ - Y))
    m=X.shape[0]
    return grad/m
def gradient_descent(X,Y,learning_rate=0.00001,max_epochs=1):
    n=X.shape[1]
    print(n)
    theta=np.zeros((n,))
    print(theta)
    error_list=[]
    print(error_list)
    for i in range(max_epochs):
        
        e=error(X,Y,theta)
        error_list.append(e)
        
        grad=gradient(X,Y,theta)
        theta=theta-learning_rate*grad
    return theta,error_list
```

```python
theta,error_list=gradient_descent(X,Y)
print(theta)
```
