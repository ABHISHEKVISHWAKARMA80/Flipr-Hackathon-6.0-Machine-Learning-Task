```python
xx = pd.read_csv('new_train_data_part02.csv').values
xx_test = pd.read_csv('test_data_for_part02.csv').values
```

```python
X=xx[:,:5]
Y=xx[:,5]
X
```

```python
Y
```

```python
std=np.std(X,axis=0)
u=np.mean(X,axis=0)
print(std)
print(u)
```

```python
X=(X-u)/std
print(X)
```

```python
ones=np.ones((X.shape[0],1))
X = np.hstack((ones,X))
print(X)
```

```python
twos=np.ones((xx_test.shape[0],1))
xx_test=np.hstack((twos,xx_test))
print(xx_test)
```

```python
df=pd.DataFrame(X)
print(df)
```

```python
def hypothesis(X,theta):
    return np.dot(X,theta)
def error(X,Y,theta):
    error=0.0
    m=X.shape[0]
    y_=hypothesis(X,theta)
    e=np.sum((Y-y_)**2)
    return e/m
def gradient(X,Y,theta):
    y_=hypothesis(X,theta)
    grad=np.dot(X.T,(y_ - Y))
    m=X.shape[0]
    return grad/m
def gradient_descent(X,Y,learning_rate=0.00001,max_epochs=1):
    n=X.shape[1]
    print(n)
    theta=np.zeros((n,))
    print(theta)
    error_list=[]
    print(error_list)
    for i in range(max_epochs):
        
        e=error(X,Y,theta)
        error_list.append(e)
        
        grad=gradient(X,Y,theta)
        theta=theta-learning_rate*grad
    return theta,error_list
```

```python
theta,error_list=gradient_descent(X,Y)
print(theta)
```

```python
print(error_list)
```

```python
y_test=hypothesis(xx_test,theta)
print(y_test)
```
