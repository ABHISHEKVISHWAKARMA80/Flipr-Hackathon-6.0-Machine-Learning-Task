```python
def change_type(x):
    if x=='C-1T':
        return 1
    if x=='CB':
        return 2
    if x=='CMC':
        return 3
    if x=='CT':
        return 4
    if x=='M':
        return 5
    if x=='MB':
        return 6
    if x=='MC':
        return 7
    if x=='MCL':
        return 8
    if x=='MCorP':
        return 9
    if x=='MPUA':
        return 10
    if x=='NAC':
        return 11
    if x=='NP':
        return 12
    if x=='NPD':
        return 13
    if x=='NPP':
        return 14
    if x=='NT':
        return 15
    if x=='NagarParishaD':
        return 16
    if x=='T':
        return 17
    if x=='TC':
        return 18
    if x=='TMC':
        return 19
    if x=='TP':
        return 20
    if x=='UA':
        return 21
```

```python
x_test['Type'] = x_test['Type'].apply(change_type)
x_test['State'] = x_test['State'].apply(change_state)
x_test
```

```python
s_imputer = SimpleImputer(missing_values=np.nan,strategy='mean',verbose=0)
```

```python
x_test = x_test.values
X=x_test[:,1:15]
s_imputer = s_imputer.fit(X)
X = s_imputer.transform(X)
print(X)
```

```python
df=pd.DataFrame(data=X,columns=["State","Type","Population","Sex Ratio","Median Age","Avg Temp","SVM","Toilets Avl","Water Purity","H Index","Female Population","# of hospitals","Foreign Visitors"])
df.to_csv("new_test_data.csv",index=True)
```
